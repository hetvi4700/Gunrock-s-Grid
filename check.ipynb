```python
import pandas as pd

# Load the enriched listings dataset
df = pd.read_csv("./enriched_listings.csv")

print(f"Total rows before dropping: {len(df)}")

# Drop rows with missing essential columns
df_cleaned = df.dropna(subset=["lat", "lon", "complex_name", "price_total", "bedrooms", "price_per_bed"]).copy()

print(f"Total rows after dropping: {len(df_cleaned)}")
print(f"Rows dropped: {len(df) - len(df_cleaned)}")
```

```python
bad_rows = df_cleaned[pd.to_numeric(df_cleaned["price_per_bed"], errors="coerce").isna()]
```
